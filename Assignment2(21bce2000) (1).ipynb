### Data visualisation

```python
sns.scatterplot(x="total",y="alcohol",data=df)
```

## The graph infers that the total accidents that have taken place are directly proportional to the amount of alcohol consumed.



```python
sns.scatterplot(y="total",x="not_distracted",data=df)
```

## Here the graph infers that the total accidents that have taken place are directly proportional to the accidents that have taken place because of not being distracted.

```python
sns.lineplot(x="total",y="no_previous",data=df)
```

## Here the graph infers that the total accidents that have taken place are directly proportional to accidents that not taken place prior.


```python
sns.relplot(x="total",y="alcohol",data=df,hue="speeding")
```

## Here the graph infers that the total accidents that have taken place due to consumption of alcohol are directly proportional to speed with which the car was driven

```python
sns.barplot(x="not_distracted",y="total",data=df,ci=None)
```

## The graph infers that the total accidents that have taken place are directly proportional to the reason being that the driver was not distracted

```python
sns.countplot(x="ins_premium",data=df)
```

## The graph gives us the count of total ins_premium





```python
sns.jointplot(x="ins_losses",y="total",data=df)
```

## The above graph gives data plotted in two differnt types.

```python
sns.boxplot(x="total",y="alcohol",data=df)
```

## Here the graph infers that the total accidents that have taken place are directly proportional to the amount of alcohol consumed
